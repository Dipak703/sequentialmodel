```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Tata-steel 2.csv')
x = pd.get_dummies(df.drop([ 'Spread Close-Open','Date'], axis=1))
y = df[ 'Spread Close-Open' ].apply(lambda x:1 if x>0 else 0 )
xtrain,xtest,ytrain,ytest = train_test_split(x,y)
```

```python
df.head()
```

```python
model = Sequential()
model.add(Dense(units=64,activation='tanh',input_dim=len(xtrain.columns)))
model.add(Dense(units=32,activation= "tanh"))

model.add(Dense(units=1,activation= 'sigmoid'))
y.head()
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(xtrain,ytrain, epochs=500,batch_size=35)
```

```python
y_hat = model.predict(xtest)
y_hat = [0 if val > 0.5 else 1 for val in y_hat]
```

```python
# mean_squared_error(y_hat,ytest)
accuracy_score(ytest,y_hat)
```
